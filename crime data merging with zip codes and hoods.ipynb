####  https://data.cityofnewyork.us/Public-Safety/NYPD-Complaint-Data-Current-Year-To-Date-/5uac-w243

```python
data=pd.read_csv("NYPD_Complaint_Data_Current__Year_To_Date_.csv")
```

```python
#adding geo = merged lat and long columns to merge wth other data frame
data["Latitude"]=data["Latitude"].astype(str)
data["Longitude"]=data["Longitude"].astype(str)
data["geo2"]=data.apply(lambda row: row.Latitude[0:5]+","+row.Longitude[0:6], axis=1)
data["geo4"]=data.apply(lambda row: row.Latitude[0:7]+","+row.Longitude[0:8], axis=1)
data["geo3"]=data.apply(lambda row: row.Latitude[0:6]+","+row.Longitude[0:7], axis=1)
```
