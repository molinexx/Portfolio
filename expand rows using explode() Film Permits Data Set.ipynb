```python
movies=pd.read_csv("Film_Permits.csv")
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.dropna(subset=["ZipCode"])
```

```python
movies.info()
```

```python
movies["ZipCode"]=movies["ZipCode"].astype(str)
```

```python
movies["zipcodes"]=movies.apply(lambda row: row.ZipCode.split(",") if len(row.ZipCode)>5 else row.ZipCode , axis=1)
```

```python
new_movies=movies.explode("zipcodes")
```

```python
new_movies.head(30)
```

```python
new_movies.to_csv("NewFilmPermits.csv")
```
