```python
import pandas as pd
```

### https://www.kaggle.com/muonneutrino/new-york-city-census-data

```python
cansus=pd.read_csv("census_block_loc.csv")
```

```python
#creating Census Track from BlockCode
cansus["BlockCode"]=cansus["BlockCode"].astype(str)
cansus["CensusTract"]=cansus.apply(lambda row: row.BlockCode[0:11], axis=1)

cansus.head()
```

```python
cansus.shape
```

```python
#drop duplicates to avoid creating cartesian product while merging
new_cesus=cansus.drop_duplicates(subset="CensusTract")
```

```python
new_cesus.shape
```

```python
#renaming for consistency
new_cesus.rename(columns={"TRACT":"CensusTract"})
```

```python
# getting census data from the same source as above
```

```python
census_nyc=pd.read_csv("nyc_census_tracts.csv")
```

```python
census_nyc.head()
```

```python
census_nyc.shape
```

```python
census_nyc["CensusTract"]=census_nyc["CensusTract"].astype(str)
cesnus_with_lat = pd.merge(census_nyc, new_cesus, on='CensusTract', how="inner")
```

```python
# after merging we lost a couple of records but no cartesian product
cesnus_with_lat.shape
```

```python
cesnus_with_lat.head()
```

```python
zips_and_cesus=zips_and_cesus[['ZIP', 'TRACT']]
```

```python
#beacuse censu track in new_zip_and-cenus is integer we need to change it to str, to match the one in second data set
zips_and_cesus["TRACT"]=zips_and_cesus["TRACT"].astype(str)
```

```python
new_zip_and_census=zips_and_cesus.drop_duplicates(subset="TRACT")
```

```python
merged_census_final=pd.merge(cesnus_with_lat,  new_zip_and_census,  left_on="CensusTract", right_on="TRACT")
```

```python
#merged resulted in few less rows, but at least without cartesian product
merged_census_final.shape
```

```python
merged_census_final.head(15)
```

```python
#add geo for connection later on
```

```python
merged_census_final["Latitude"]=merged_census_final["Latitude"].astype(str)
merged_census_final["Longitude"]=merged_census_final["Longitude"].astype(str)

merged_census_final["geo4"]=merged_census_final.apply(lambda row: row.Latitude[0:7]+","+row.Longitude[0:8], axis=1)
merged_census_final["geo3"]=merged_census_final.apply(lambda row: row.Latitude[0:6]+","+row.Longitude[0:7], axis=1)
merged_census_final["geo2"]=merged_census_final.apply(lambda row: row.Latitude[0:5]+","+row.Longitude[0:6], axis=1)
```
