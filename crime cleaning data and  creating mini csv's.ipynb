```python
import pandas as pd
```

```python
dt=pd.read_csv("merged_crime_zip_hood.csv")

dt.head()
```

```python
dt["State"]="NY"
```

```python
col_list=list(dt.columns)
print(col_list)
```

```python
dt=dt[['CMPLNT_NUM', 'CMPLNT_FR_DT', 'Month', 'Year', 'CMPLNT_FR_TM', 'BORO_NM','State', 'Incident Zip', 'neighbourhood', 
        'SUSP_AGE_GROUP', 'SUSP_RACE', 'SUSP_SEX', 'VIC_AGE_GROUP', 'VIC_RACE', 'VIC_SEX', 
         'LAW_CAT_CD', 'OFNS_DESC', 'CRM_ATPT_CPTD_CD']]
```

```python
dt.tail(10)
```

```python
col_list=list(dt.columns)
print(col_list)
```

```python
dt["Year"].unique()
```

```python
dt.info()
```

```python
dt["Year"]=dt["Year"].astype(str)
```

```python
dt.drop(dt[dt["Year"]!="2019"].index, inplace=True)
```

```python
dt["Year"].unique()
```

```python
dt.head()
```

```python
dt=dt.dropna(subset=["Incident Zip"])
```

```python
dt["Incident Zip"]=dt["Incident Zip"].astype(int)
```

```python
dt["Incident Zip"]=dt["Incident Zip"].astype(str)
```

```python
 #we have 13,216 empty rows in col zipcodes but only 6,486 mising in "neighbourhood"

null_columns=dt.columns[dt.isnull().any()]
dt[null_columns].isnull().sum()
```

```python
dt=dt.dropna(subset=["neighbourhood"])
```

```python
dt=dt.dropna(subset=["OFNS_DESC"])
```

```python
dt=dt.dropna(subset=["neighbourhood"])
```

```python
dt.head(20)
```

```python
# cleaning some other attributes like age
```

```python
dt["SUSP_AGE_GROUP"].unique()
```

```python
dt.groupby(["SUSP_AGE_GROUP"]).agg({"CMPLNT_NUM":"count"})
```

```python
drop_index=dt[(dt["SUSP_AGE_GROUP"]!="45-64") & (dt["SUSP_AGE_GROUP"]!="25-44")&
                    (dt["SUSP_AGE_GROUP"]!="18-24") & (dt["SUSP_AGE_GROUP"]!="65+")&
                   (dt["SUSP_AGE_GROUP"]!="<18")& (dt["SUSP_AGE_GROUP"]!="UNKNOWN")].index

dt.drop(drop_index , inplace=True)
dt["SUSP_AGE_GROUP"].unique()
```

```python
drop_index=dt[(dt["VIC_AGE_GROUP"]!="45-64") & (dt["VIC_AGE_GROUP"]!="25-44")&
                    (dt["VIC_AGE_GROUP"]!="18-24") & (dt["VIC_AGE_GROUP"]!="65+")&
                   (dt["VIC_AGE_GROUP"]!="<18")& (dt["VIC_AGE_GROUP"]!="UNKNOWN")].index

dt.drop(drop_index , inplace=True)
dt["SUSP_AGE_GROUP"].unique()
```

```python
dt.info()
```

```python
dt["Month"]=dt["Month"].astype(str)
```

```python
dt["CMPLNT_NUM"]=dt["CMPLNT_NUM"].astype(str)
```

```python
dt.info()
```

```python
dt.rename(columns={"CRM_ATPT_CPTD_CD":"CrimeSuccesOrNot", "CMPLNT_FR_DT":"DateOfCrime", 
                     "CMPLNT_FR_TM":"TimeOfCrime" , "LAW_CAT_CD":"CrimeCategory", "OFNS_DESC":"CrimeDescription"})
```

```python
dt.to_csv("FinalCrime.csv")
```
